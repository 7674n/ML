```python
import pandas as pd
import numpy as np
```

```python
# dataframes for data inconsistency practice
df1 = {
    'date': ['2019-01-01', '2019/01/02', '2019-12-15', '04-01-2019'],
    'name': ['amaar', 'amar', 'hamza', 'hazma'],
    'country': ['USA', 'U.S.A', 'America', 'United States of America'],
    'sale_2019': [100, 200, None, 400],
    'sale_2020': [None, 210, 310, 410]
}

df1 = pd.DataFrame(df1)
```

```python
df1.head()
```

```python
# dataframes for date inconsistency to  replace with correct dateformat
df1['date'] = pd.to_datetime(df1['date'], errors='coerce')
df1['date'] = df1['date'].dt.strftime('%Y-%m-%d')
df1.head()
```

from datetime import datetime
import pandas as pd

```python
country_mapping = { 'USA': 'United States of America', 'U.S.A': 'United States of America', 'America': 'United States of America' }
df1['country'] = df1['country'].replace(country_mapping)
df1.head()
```

```python
# name inconsistency imuute with correct name
df1['name'] = df1['name'].str.capitalize()
#replace
df1['name'] = df1['name'].replace({'Hazma': 'Hamza', 'Amar': 'Amaar'})
df1.head()
```

```python
#drop duplicates
df1.drop_duplicates(subset=['name'], inplace=True)
df1.head()
```
