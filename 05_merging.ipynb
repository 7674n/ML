# Data Merging

```python
import pandas as pd
import numpy as np
```

```python
data1= {"id" : [1,2,3,4,],
        'name':['Asad','Ali','Umer','dail'],
        'age':[10,20,30,40]
}

df1 = pd.DataFrame(data1)
df1       
```

```python
data2= {"id" : [1,2,3,4,],
        'city':['lahore','karachi','islamabad','peshawar'],
        'country':['pakistan','pakistan','pakistan','pakistan']
}

df2 = pd.DataFrame(data2)
df2
```

```python
# merge  data sets
merg_df=pd.merge(df1,df2,on='id', how='inner')
merg_df
```

# concat

```python
con=pd.concat([df1,df2],axis=1)
con
```

```python
con1=pd.concat([df1,df2],axis=1)
con1
```
